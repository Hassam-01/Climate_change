```python
split_index = int(0.8 * len(flood_df))
train_df = flood_df[:split_index]
test_df = flood_df[split_index:]

x_train = train_df.drop("FloodProbability", axis=1)
y_train = train_df["FloodProbability"]

X_test = test_df.drop("FloodProbability", axis=1)
Y_test = test_df["FloodProbability"]
```

```python
X_test.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

model = RandomForestRegressor()
model.fit(x_train, y_train)

y_pred = model.predict(X_test)

mse = mean_squared_error(Y_test, y_pred)
r2 = r2_score(Y_test, y_pred)

mse, r2
```

```python
from sklearn.model_selection import train_test_split, RandomizedSearchCV, KFold
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

# 1) Split once; keep test set locked away
# X_train, X_test, y_train, y_test = train_test_split(train_df, test_df, test_size=0.2, random_state=42)

# 2) Define model & search space
rf = RandomForestRegressor(random_state=42)

param_dist = {
    "n_estimators": [100, 200, 400],  # more trees => stabler, but slower
    "max_depth": [10, 20, None],  # limit depth to prevent overfitting
    "min_samples_split": [2, 10, 20],  # how strict split is
    "min_samples_leaf": [1, 2, 4],
    "bootstrap": [True, False],
}

cv = KFold(n_splits=5, shuffle=True, random_state=42)

search = RandomizedSearchCV(
    rf,
    param_distributions=param_dist,
    n_iter=10,  # budget: try ~60 combos
    scoring="r2",  # or "neg_root_mean_squared_error"
    cv=cv,
    n_jobs=1,
    verbose=1,
    random_state=42,
)

# 3) Fit on training only (CV happens inside)
search.fit(x_train, y_train)

best_rf = search.best_estimator_
print("Best params:", search.best_params_)
print("CV best R2:", search.best_score_)

# 4) Final holdout evaluation
pred = best_rf.predict(X_test)
print("Test R2:", r2_score(Y_test, pred))
print("Test MAE:", mean_absolute_error(Y_test, pred))
print("Test RMSE:", mean_squared_error(Y_test, pred, squared=False))
```
